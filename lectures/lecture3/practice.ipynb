```python
import torch
```

```python
data = open('names.txt','r').read().split()
data[:8]
```

```python
len(data)
```

```python
#building a vocabulary
vocab = list(set(''.join(data)))
len(vocab)
```

```python
vocab.sort()
vocab
```

```python
itos = {0:'.'}
for i,x in enumerate(vocab):
    itos[i+1] = x
itos
```

```python
stoi = {}
for i,j in itos.items():
    stoi[j]= i
stoi
```

```python
ab = [0,0,0]
ab = ab[1:] + [5]
ab
```

```python

X = []
Y = []
for word in data:
    context = [0]*6
    for c in word:
        X.append(context)
        Y.append(stoi[c])
        ix = stoi[c]
        context = context[1:] + [ix]
    X.append(context)
    Y.append(0)

```

```python
for i in range(50):
    print(X[i])
    print(Y[i])
```

```python
Y
```

```python
X = torch.tensor(X)
Y = torch.tensor(Y)
```

```python
X.shape, Y.shape
```

```python
# now we have the input data and the target data , we will try to one hot encode the input data
one_hot = torch.nn.functional.one_hot(X, num_classes=27)
one_hot = one_hot.view(228146*6,27).float()
one_hot.shape
```

```python
# now we will build the weight matrix for the embedding layer
C = torch.randn(27, 3)
C
```

```python
emb = one_hot @ C
emb.shape
```

```python
W1 = torch.randn((3,100))
b1 = torch.randn(100)

h = emb @ W1 + b1
h
```

```python
h.shape
```

```python
h = torch.tanh(h)
```

```python
h.shape
```

```python
W2 = torch.randn((100, 27))
b2 = torch.randn(27)
logits = h @ W2 + b2
```

```python
logits.shape
```

```python
counts = logits.exp()
prob = counts / counts.sum(1, keepdim=True)
```

```python
prob.shape
```

```python
prob[1].sum()
```

```python
parameters = [C, W1, b1, W2, b2]
sum(p.nelement() for p in parameters) 
```

```python
for i in range(1000):
    emb = one_hot @ C
    h = emb @ W1 + b1
    h = torch.tanh(h)
    logits = h @ W2 + b2
    loss = torch.nn.functional.cross_entropy(logits, Y)

    for p in parameters:
        p.grad = None
        loss.backward()

    
    
```
